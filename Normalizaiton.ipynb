**NORMALIZING THE DATA USING DIFFERENT TECHNIQUES**

```python
import pandas as pd
```

```python
#Create a data
df = pd.DataFrame(
      [
          [25000,200,30.5],
           [18000,150,12.5],
           [9000,100,25.7],
            [40000,300,21.2]
       ],
           columns = ["A", "B","C"]
)
```

```python
# To display data
display(df)
```

```python
#visualize the data and set the plot
import matplotlib.pyplot as plt
df.plot(kind = 'bar', )
```

Here, A is dominant and we can't see B and C. So, we need to normalize
      ** Maximum Absolute Scaling **

```python
#Don't bother original data and need to create a copy of that and do operations
df_copy = df.copy()
```

```python
df_copy
```

```python
#Apply Maximum Absolute Scaling
for col in df_copy.columns:
    df_copy[col] = df_copy[col] / df_copy[col].abs().max()
```

```python
#Display data to see the rescaled info...
df_copy
```

```python
#To see maximum & absolute value in the above data
df.abs().max()
```

```python
df_copy.plot(kind = 'bar')
```

**second technique using normalizing data i.e., Min-Max Feature Scaling or Normalization**

```python
df_copy_minmax = df.copy()
```

```python
df_copy_minmax
```

```python
for col in df_copy_minmax.columns:
    df_copy_minmax[col] =  (df_copy_minmax[col] - df_copy_minmax[col].min()) / (df_copy_minmax[col].max() - df_copy_minmax[col].min())
```

```python
df_copy_minmax
```

```python
#Visualize and see
df_copy_minmax.plot(kind = 'bar')
```

** third technique Z score method or standardisation**

```python
#Need to copy of data
df_copy_zscore = df.copy()
```

```python
df_copy_zscore
```

```python
for col in df_copy_zscore.columns:
    df_copy_zscore[col] = (df_copy_zscore[col] - df_copy_zscore[col].mean()) / df_copy_zscore[col].std()
```

```python
df_copy_zscore
```

```python
#visualize and see
df_copy_zscore.plot(kind = 'bar')
```
